```python
doc = cnlp.Document(xmls[37])
model = RushModel(doc, history=2)
trans = model.gold_transitions()
order = recover_order(model.gold_transitions())

G = pgv.AGraph()
nsents = len(order)
for i, sindex in enumerate(order):
    for j, w in enumerate(doc[sindex]):
        G.add_node(u'{}x{} {}'.format(sindex, j, unicode(w)), shape='box', fontsize='10.0', pos='{},{}!'.format(j*1.5, nsents - i))


for tran in trans:
    model.feature_map(tran)
    feats = model._t
    for feat in feats:
        #print feat
        indices = feat[1]
        s0 = indices[0][0]
        w0 = indices[0][1]
        s1 = indices[1][0]
        w1 = indices[1][1]
        n0 = u'{}x{} {}'.format(s0, w0, unicode(doc[s0][w0]))
        n1 = u'{}x{} {}'.format(s1, w1, unicode(doc[s1][w1]))
        
        G.add_edge(n0, n1, label=feat[0])
    
G.layout(prog='neato')
G.draw('/tmp/toctree.png')
from IPython.display import Image
Image(filename='/tmp/toctree.png')
```
