```python
def paths_to_perm(perm, constraints=(), nconstraints=()):
    for i, x in enumerate(perm):
        if x == 1:
            clone = list(perm)
            clone[i] = 0
            feasible = True
            for con in constraints:
                if clone[con] != 1:
                    feasible = False
                    break
            for con in nconstraints:
                if clone[con] != 0:
                    feasible = False
                    break
            if feasible:
                yield clone
    
    
list(paths_to_perm([0,1,1,1,0,1], nconstraints=[2]))
```
